Jointplot is used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sns.jointplot(x='SepalLengthCm',y='SepalWidthCm',data=iris)
plt.show()
```

We can identify outliers using the regression line on a scatter plot.

```python
sns.jointplot("SepalLengthCm","SepalWidthCm",data=iris,kind="reg")
plt.show()
```

Combining Stripplot and boxplot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='SepalLengthCm',data=iris)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
ax= sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax= sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")

boxtwo = ax.artists[2]
boxtwo.set_facecolor('yellow')
boxtwo.set_edgecolor('black')
boxthree=ax.artists[1]
boxthree.set_facecolor('red')
boxthree.set_edgecolor('black')
boxthree=ax.artists[0]
boxthree.set_facecolor('green')
boxthree.set_edgecolor('black')

plt.show()
```

```python
iris['Species']=iris['Species'].astype('category')
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)

h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```
